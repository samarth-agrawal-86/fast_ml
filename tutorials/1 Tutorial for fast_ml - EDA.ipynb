# Tutorial for using the package `fast-ml` 

This package is as good as having a junior Data Scientist working for you. Most of the commonly used EDA steps, Missing Data Imputation techniques, Feature Engineering steps are covered in a ready to use format

## Part 1. EDA for Numerical Variables & Categorical Variables


#### Available functions inside the eda module
#### For Numerical Variable
> * eda_numerical_variable
> * eda_numerical_plots
> * eda_numerical_plots_with_target

#### For Categorical Variable
> * eda_categorical_variable
> * eda_category_plots
> * eda_category_plots_with_target

#### 1. From eda module import function for numerical variable and categorical variable
`from fast_ml.eda import eda_numerical_variable, eda_categorical_variable`

#### 2. Select a variable and run the command for various exploratory analysis
* For Categorical variables use `eda_categorical_variable`
* For Numerical variables use `eda_numerical_variable`

`eda_numerical_variable(df, variable='MSSubClass', model = 'reg', target = 'SalePrice', threshold=10)`

#### 3. To get Category plots for all the variables import function for plotting
`from fast_ml.eda import eda_category_plots, eda_category_plots_with_target`

#### 4. Run the following command to get all the charts
This takes input the list of all the categorical variables you want to plot<br>
`eda_category_plots(df, variables, add_missing = True, add_rare = False, tol=0.5)`


#### 5. To get Numerical plots for all the variables import function for plotting
`from fast_ml.eda import eda_numerical_plots_with_target`

#### 6. Run the following command to get all the charts
This takes input the list of all the numerical variables you want to plot<br>
`eda_numerical_plots_with_target(df, variables, target, model = 'clf')`
