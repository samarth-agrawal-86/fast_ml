# Tutorial for using the package `fast-ml` 

This package is as good as having a junior Data Scientist working for you. Most of the commonly used EDA steps, Missing Data Imputation techniques, Feature Engineering steps are covered in a ready to use format

## Part 1. EDA for Numerical Variables & Categorical Variables



#### 1. Import eda module from the package 
`from fast_ml.eda import eda`

#### 2. Define the eda object. 
* If you want to see the relationship with target variable then provide the target varaible
* Provide what kind of problem you are solving - regression / classification

`eda_obj = eda(df, target=, model = )`

#### 3. Select a variable and run the command for various exploratory analysis
* For Categorical variables use `eda_categorical_variable`
* For Numerical variables use `eda_numerical_variable'

`eda_obj.eda_categorical_variable('MSSubClass')`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from fast_ml.eda import eda

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')
df.shape
```

```python
df.head(5)
```

```python
numeric_type = ['float64', 'int64']
category_type = ['object']
```

```python
# All Numerical Variables
df.select_dtypes(include=numeric_type).columns
```

```python
# All Categorical Variables
df.select_dtypes(include=category_type).columns
```

## Start EDA

```python
# Define EDA Object
eda_obj = eda(df, target='SalePrice', model = 'regression')
```

## Numerical Variables 

### 1. MSSubClass

```python
# Use the following method for a numerical variable 
eda_obj.eda_numerical_variable('MSSubClass')
```

### 2.  LotFrontage

```python
# Use the following method for a numerical variable 
eda_obj.eda_numerical_variable('LotFrontage')
```

## Categorical Variables

### 1. MSZoning 

```python
# Use the following method for a categorical variable 
eda_obj.eda_categorical_variable('MSZoning')
```

### 2. Street

```python
# Use the following method for a categorical variable 
eda_obj.eda_categorical_variable('RoofStyle')
```
